# train과 test 범주형 데이터 비교
object 컬럼과 카테고리가 많이 있을때 train과 test 차이를 눈으로 비교하기 어렵다
또한 카테고리 개수 같더라도 값이 다를 수 있다. 중복을 허용하지 않는 set을 통해 비교하자.

```python
cols = train.select_dtypes(include='object').columns
for col in cols:
  set_train = set(train[col])
  set_test = set(test[col])
  same = (set_train == set_test)
  if same:
    print(col,"\t 동일함")
  else:
    print(col,"\t 동일하지않음")
```

```python
target = train.pop('target')
```

```python
# 결측치를 모두 X 처리

train = train.fillna('X')
test = test.fillna('X')

combined = pd.concat([train,test])
combined_dummies = pd.get_dummies(combined)
n_train = len(train)
train = combined_dummies[:n_train]
test = combined_dummies[n_train:]
```

```python
from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(train,target,test_size=0.2,random_state=0)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(X_tr,y_tr)
pred = rf.predict_proba(X_val)

from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(y_val,pred[:,1])
print('roc-auc: ',roc_auc)

pred = rf.predict_proba(test)
submit = pd.DataFrame({'pred' :pred[:,1]})
submit.to_csv("result.csv",index=False)
```

#### 성능개선
    원-핫 인코딩 -> 레이블 인코딩 변경
    스케일링
    id 제거

```python
target = train.pop('target')
train = train.fillna('X')
test = test.fillna('X')
from sklearn.preprocessing import LabelEncoder
combined = pd.concat([train,test])
cols = train.select_dtypes(include='object').columns
```

```python


for col in cols:
  le = LabelEncoder()
  combined[col] = le.fit_transform(combined[col])

n_train = len(train)
train = combined[:n_train]
test = combined[n_train:]

from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
n_cols = train.select_dtypes(exclude='object').columns
train = scaler.fit_transform(train)
test = scaler.fit_transform(test)

from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(train,target,test_size=0.2,random_state=0)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=5,n_estimators=500,random_state=0)
rf.fit(X_tr,y_tr)
pred = rf.predict_proba(X_val)

from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(y_val,pred[:,1])
print("roc_auc: ",roc_auc)

# 성능향상
pred = rf.predict_proba(test)
submit = pd.DataFrame({'pred':pred[:,1]})
submit.to_csv("result.csv",index=False)
```

#### 컬럼 삭제 전 반드시 전후 크기를 비교하자

```python
print("삭제 전 ",train.shape)
train.dropna(subset=['OCCUPATION_TYPE'],inplace=True)
print("삭제 후",train.shape)
```

```python
target = train.pop('STATUS')
```

```python
# 원-핫 인코딩
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

```python
from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(train,target,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(X_tr,y_tr)
pred = rf.predict(X_val)
```

```python
from sklearn.metrics import f1_score
score = f1_score(y_val,pred)
print("F1 Score: ",score)
```

```python
pred = rf.predict(test)
submit = pd.DataFrame({'pred':pred})
submit.to_csv("result.csv",index=False)
```

#### 성능개선
    불균형 데이터일 경우 평가지표가 낮게 나올수 있다. 현재 F1_SCORE = 0.22
    F1 스코어는 1에 가까울소록 좋은 성능이다.
    불균형 데이터를 자동 균형있게 조절하는 -> class_weight='balanced'

```python
# 결측치 제거
freq = train['OCCUPATION_TYPE'].mode()[0]
train['OCCUPATION_TYPE'].fillna(freq)
test['OCCUPATION_TYPE'].fillna(freq)
```

```python
target = train.pop('STATUS')
```

```python
train.head()
```

```python
train = train.drop('ID',axis=1)
test = test.drop('ID',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
cols = train.select_dtypes(include='object').columns
for col in cols:
  le = LabelEncoder()
  train[col] = le.fit_transform(train[col])
  test[col] = le.fit_transform(test[col])
```

```python
from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(train,target,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier(n_estimators=500,class_weight='balanced',random_state=0)
rf.fit(X_tr,y_tr)
pred = rf.predict(X_val)
```

```python
from sklearn.metrics import f1_score
score = f1_score(y_val,pred)
print(score)

pred = rf.predict(test)
submit = pd.DataFrame({'pred':pred})
submit.to_csv("result.csv",index=False)
```
