```python
import pandas as pd
df = pd.read_csv("delivery_time.csv")
# 배달료 계산 기준표에 따라 각 주문에 대한 배달료를 계산하시오
# (5km 미만 : 2000원 / 5km 이상 10km 미만 : 4000원 / 10km 이상 15km 미만 6000원 / 15km 이상 20km 미만 8000원)
def delivery_fee(distance):
  if(distance <5):
    return 2000
  elif(distance<10):
    return 4000
  elif(distance<15):
    return 6000
  elif(distance<20):
    return 8000
df['배달료'] = df['거리'].apply(delivery_fee)
df['배달료']
# 월별로 배달료의 총합을 집계하시오
df['주문시간'] = pd.to_datetime(df['주문시간'])
monthData = df['주문시간'].dt.to_period('M')
month_sum = df.groupby(monthData)['배달료'].sum()
month_sum
# 배달료가 가장 많이 발생한 월을 찾고, 그 월의 총 배달료를 정수로 구하시오.
max_fee_month = month_sum.idxmax()
max_fee_value = month_sum[max_fee_month]
print(max_fee_value)
```
