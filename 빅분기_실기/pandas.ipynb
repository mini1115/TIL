```python
listbox = ["감사","행복","사랑","성공","긍정","변화","성장","희망"]

# 리스트의 있는 값의 수를 출력하시오
print(len(listbox))
# 리스트의 첫번쨰 값을 출력하시오
print(listbox[0])
# 리스트의 마지막 값을 출력하시오
print(listbox[-1])
# 리스트의 처음부터 세번째 값까지 출력하시오
print(listbox[:3])
# 리스트에서 마지막 앞 단어를 출력하시오
print(listbox[-2])
# 리스트 값중 '행복'과 '사랑'만 출력하시오
for item in listbox:
  if(item=='행복'or item=='사랑'):
    print(item)
# 리스트 값 중 '변화'->'웃음'으로 변경하시오
listbox[-3] = "웃음"
print(listbox)
```

```python
listbox = [15,46,78,24,56]

def max_min_diff(data):
  ma = max(data)
  mi = min(data)
  return ma-mi

print(max_min_diff(listbox))
```
