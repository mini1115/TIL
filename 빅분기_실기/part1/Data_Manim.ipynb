Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 Unknown : N
Less than $40K : a
$40K - $60K : b
$60K - $80K : c
$80K - $120K : d
$120K +’ : e

```python
def change(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':
        return 'b'
    elif x =='$60K - $80K':
        return 'c'
    elif x =='$80K - $120K':
        return 'd'
    elif x =='$120K +' :
        return 'e'

df['newIncome'] = df['Income_Category'].apply(change)
df['newIncome']
```

Map방식 변

```python
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'
}

df['newIncome'] = df['Income_Category'].map(lambda x:dic[x])
df['newIncome']
```

Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라. (0~9 : 0 , 10~19 :10 , 20~29 :20 … 각 구간의 빈도수를 출력하라

```python
def convert(x):
  if x<10:
    return 0
  elif x<19:
    return 10
  elif x<29:
    return 20
  elif x<39:
    return 30
  elif x<49:
    return 40
  elif x<59:
    return 50
  elif x<69:
    return 60
  elif x<79:
    return 70
```

```python
df['AgeState'] = df['Customer_Age'].apply(convert)
df['AgeState'].value_counts().sort_index()
```

년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다

```python
# 24시 -> 00시 + 하루 추가
df['(년-월-일:시)']
df.rename(columns= {'(년-월-일:시)':'date'})
```

```python
def change(x):
  import datetime
  hour = x.split(":")[1]
  date = x.split(":")[0]

  if hour == '24':
    hour = '00:00:00'

    FinalDate = pd.to_datetime(date+" "+hour) + datetime.timedelta(days=1)

  else:
    hour = hour +':00:00'
    FinalDate = pd.to_datetime(date+" "+hour)

  return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change)
df['(년-월-일:시)']
```
