Q. 다음은 특정 질병(disease)의 유무를 나타내는 환자 데이터셋이다.

```python
import pandas as pd
df = pd.read_csv('health_survey.csv')
df.head(3)
```

- age와 bmi 지수를 독립변수로 활용해 질병(disease)의 발생 여부를 예측

```python
from statsmodels.formula.api import logit
import statsmodels.api as sm

model = logit('disease ~age +bmi',data=df).fit()
print(model.summary())
```

- bmi 변수가 한 단위 증가 시 질병 발생의 오즈비(Odds Ratio)값은?

```python
import numpy as np
print(model.params['bmi'])
print(np.exp(model.params['bmi']))
```

```python
# 로그우도
print("Log-Likelihodd :",model.llf)
print("Deviance: ",model.llf*-2)
```
