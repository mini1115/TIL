```python
# 일원 분산분석
# 4개의 중학교에서 다른 교육 방법을 사용해 수학을 학생들에게 가르쳤다.
# 각 중학교 학생들이 받은 교육 후 수학 성적은 다음과 같다.
# 이 4가지 방법의 성적차이가 통계적으로 유의한지 판단하라(0.05 유의수준)
df = pd.read_csv('math.csv')
df.head(4)
```

#### 이원 분산 분석
    H 토마토 수학량을 늘리기 위해 세 가지 다른 비료 유형과 네가지 다른 물 주기 실험하고자 한다.
    연구소에서는 12개의 토마토 식물을 무작위로 선택해 각 조합에 대해 반복적으로 실험 수행.
-  비료 유형에 따른 유형
  - 귀무가설 : 모든 비료 유형의 토마토 수확량 평균은 동일하다.
  - 대립가설 : 적어도 2개 이상의 비료 유형에서의 토마토 수확량 평균이 다르다.

-  물 주기에 따른 유형
  - 귀무가설 : 모든 물 주기의 토마토 수확량 평균은 동일하다.
  - 대립가설 : 적어도 2개 이상의 비료 물 주기에서의 토마토 수확량 평균이 다르다.

-  비료 유형과 물 주기 간 상호작용 효과
  - 귀무가설 : 비료 유형과 물 주기 간 상호작용은 토마토 수확량에 영향을 미치지 않는다.
  - 대립가설 : 비료 유형과 물 주기 간 상호작용은 토마토 수확량에 영향을 미친다.

```python
df = pd.read_csv('tomato2.csv')
df.head(5)
```

```python
df['비료유형'].value_counts()
```

```python
df['물주기'].value_counts()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('수확량 ~C(비료유형)*C(물주기)',data=df).fit()
anova_table = sm.stats.anova_lm(model)
print(anova_table)
```

### 적합도 검정
    한 도시의 운전자 1,000명을 대상으로 교통사고 경험 수를 조사했다.
    1회 550명, 2회 250명, 3회 100명, 4회 70명, 5회 이상 : 30명이다.
    전국적으로 조사된 운전자들의 교통사고 경험 수 분포는 다음과 같다.
    1회: 60%, 2회 : 25%, 3회 : 8%, 4회 : 5%,5회 이상 : 2%
    이 도시 운전자들의 교통사고 경험 수 분포가 전국적인 분포를 따르는지 검정하시오.

| 구분     | 경험 수       | 전국적인 경향(%)       |
|------------|------------|------------|
| 1회   | 550   | 60%   |
| 2회   | 250   | 25%  |
| 3회   | 100   | 8%   |
| 4회   | 70   | 5%  |
| 5회   | 30   | 2%   |


```python
import pandas as pd
df = pd.DataFrame({'구분':['1회','2회','3회','4회','5회'],'경험 수':[550,250,100,70,30],'전국적인 경향(%)':[60,25,8,5,2]})
df
```

```python
print(30/1000)

observed = df['경험 수']
expected = df['전국적인 경향(%)']*10

from scipy.stats import chisquare
print(chisquare(observed,expected))
```

### 독립성 검정
    길벗 대학에는 빅데이터 분석기사 캠프와 정보처리기사 캠프가 있다.
    최근 두 캠프 학생들 사이에서 주최한 세미나에 참여하려면 등록을 해야한다.
    각 캠프의 세미나 등록 여부를 조사한 결과는 아래와 같다.
    캠프와 세미나 등록 여부가 독립적인지 검정하시오.

1. 교차표 형태 데이터

| 구분               | 등록함 | 등록안함 |
|--------------------|--------|----------|
| 빅데이터분석기사캠프 | 50     | 30       |
| 정보처리기사캠프   | 60     | 40       |

```python
df = pd.DataFrame({'구분':['빅분기','정처기'],'등록함':[50,60],'등록안함':[30,40]})
df
```

```python
from scipy.stats import chi2_contingency
observed = [[50,30],[60,40]]
chi2_contingency(observed)
```

```python
# 로우 데이터
df = pd.DataFrame({
    '캠프':['빅분기']*80 + ['정처기']*100,
    '등록여부':['등록함']*50+['등록안함']*30 + ['등록함']*60+['등록안함']*40
})
```

```python
df = pd.crosstab(df['캠프'],df['등록여부'])
print(chi2_contingency(df))
```

### 다중 선형회귀
    한 도시의 배달 서비스 회사에서 한 달 동안의 주문 데이터를 수집했다.
    배달 주문량은 할인율(%), 온도(C), 광고비(천 원)과 같은 다양한 요인들에 영향을 받을수 있다.
    배달 서비스 회사는 이런 요인들을 고려해 주문량을 예측하고자 한다.
    할인율, 온도, 광고비를 독립변수로 사용해 다중 선형 회귀 모델을 구축하고 다음 문제에 답하시오.

- 귀무가설 : 광고비의 회귀 계수는 0이다.(광고비가 배달 주문량에 영향을 주지 않는다.)
- 대립가설 : 광고비의 회귀 계수는 0이 아니다.(광고비가 배달 주문량에 영향을 준다.)

```python
# 데이터
import pandas as pd
df = pd.DataFrame({
    '할인율': [28, 24, 13, 0, 27, 30, 10, 16, 6, 5, 7, 11, 11, 30, 25,
            4, 7, 24, 19, 21, 6, 10, 26, 13, 15, 6, 12, 6, 20, 2],
    '온도': [15, 34, 15, 22, 29, 30, 14, 17, 28, 29, 19, 19, 34, 10,
           29, 28, 12, 25, 32, 28, 22, 16, 30, 11, 16, 18, 16, 33, 12, 22],
    '광고비': [342, 666, 224, 764, 148, 499, 711, 596, 797, 484, 986, 347, 146, 362, 642,
            591, 846, 260, 560, 941, 469, 309, 730, 305, 892, 147, 887, 526, 525, 884],
    '주문량': [635, 958, 525, 25, 607, 872, 858, 732, 1082, 863, 904, 686, 699, 615, 893,
            830, 856, 679, 918, 951, 789, 583, 988, 631, 866, 549, 910, 946, 647, 943]
})
print(df.head(3))
```

```python
data = df['할인율'].corr(df['온도'])
print(round(data,2))
```

```python
from statsmodels.formula.api import ols
model = ols('주문량 ~할인율+광고비+온도',data=df).fit()
round(model.rsquared,2)
```

```python
model.summary()
```

```python
model.params['온도']
model.pvalues['온도']
```

```python
round(model.params['Intercept'],4)
```

```python
new_data = pd.DataFrame({'할인율':[10],'온도':[20],'광고비':[500]})
result = model.predict(new_data)
result
```

```python
df['잔차'] = df['주문량'] - model.predict(df)
sum(df['잔차']**2)
```

```python
sum(model.resid**2)
```

```python
(df['잔차']**2).mean()
```

```python
(model.resid**2).mean()
```

```python
# 90 % 신뢰구간
print(model.conf_int(alpha=0.1))
```

```python
new_data2 = pd.DataFrame({'할인율':[15],'온도':[25],'광고비':[300000]})
pred = model.get_prediction(new_data2)
# result = pred.conf_int(alpha=0.1)
result = pred.summary_frame(alpha=0.1)
result
```

```python
# 독립변수 할인율과 온도를 고정한 상태에서 광고비가 배달 주문량에 영향을 주는지 가설 검정을 하시오.(유의수준 0.05)
cond  = model.pvalues['광고비'] < 0.05
if(cond):
    result='기각'
else:
    result='채택'
print(result)
```
