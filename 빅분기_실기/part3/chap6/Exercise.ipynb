```python

# 스마트폰 충전시간을 줄이기 위한 새로운 충전기를 개발했다.
# 개발된 충전기의 효과를 검정하기 위해 스마트폰 사용자 집단을 두 그룹으로 나눠 한그룹에는 새로운 충전기, 다른 한그룹에는 기존의 충전기를 제공
# 새로운 충전기가 더 빠르다고 할수있는지를 검정하시오.
import pandas as pd
df = pd.DataFrame({
    '충전기': ['New'] * 10 + ['Old'] * 10,
    '충전시간': [
        1.5, 1.6, 1.4, 1.7, 1.5, 1.6, 1.7, 1.4, 1.6, 1.5,
        1.7, 1.8, 1.7, 1.9, 1.8, 1.7, 1.8, 1.9, 1.7, 1.6
    ]
})
print(df.head(2))
```

```python
new = df['충전기'] == 'New'
old = df['충전기'] == 'Old'

from scipy import stats
stats.ttest_ind(df[new]['충전시간'],df[old]['충전시간'],alternative='less',equal_var=True)
```

```python
# 일원 분산분석
# 4개의 중학교에서 다른 교육 방법을 사용해 수학을 학생들에게 가르쳤다.
# 각 중학교 학생들이 받은 교육 후 수학 성적은 다음과 같다.
# 이 4가지 방법의 성적차이가 통계적으로 유의한지 판단하라(0.05 유의수준)
df = pd.read_csv('math.csv')
df.head(4)
```

```python
df['groups'].value_counts()
```

```python
groupA = df['groups']=='group_A'
groupB = df['groups']=='group_B'
groupC = df['groups']=='group_C'
groupD = df['groups']=='group_D'

from scipy import stats
stats.shapiro(df[groupA]['scores'])
stats.shapiro(df[groupB]['scores'])
stats.shapiro(df[groupC]['scores'])
stats.shapiro(df[groupD]['scores'])

print(stats.levene(df[groupA]['scores'],df[groupB]['scores'],df[groupC]['scores'],df[groupD]['scores']))
```

```python
from statsmodels.formula.api import ols
model = ols('scores ~groups',data=df).fit()

from statsmodels.stats.anova import anova_lm
anova_lm(model)
```

#### 이원 분산 분석
    H 토마토 수학량을 늘리기 위해 세 가지 다른 비료 유형과 네가지 다른 물 주기 실험하고자 한다.
    연구소에서는 12개의 토마토 식물을 무작위로 선택해 각 조합에 대해 반복적으로 실험 수행.
-  비료 유형에 따른 유형
  - 귀무가설 : 모든 비료 유형의 토마토 수확량 평균은 동일하다.
  - 대립가설 : 적어도 2개 이상의 비료 유형에서의 토마토 수확량 평균이 다르다.

-  물 주기에 따른 유형
  - 귀무가설 : 모든 물 주기의 토마토 수확량 평균은 동일하다.
  - 대립가설 : 적어도 2개 이상의 비료 물 주기에서의 토마토 수확량 평균이 다르다.

-  비료 유형과 물 주기 간 상호작용 효과
  - 귀무가설 : 비료 유형과 물 주기 간 상호작용은 토마토 수확량에 영향을 미치지 않는다.
  - 대립가설 : 비료 유형과 물 주기 간 상호작용은 토마토 수확량에 영향을 미친다.

```python
df = pd.read_csv('tomato2.csv')
df.head(5)
```

```python
df['비료유형'].value_counts()
```

```python
df['물주기'].value_counts()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('수확량 ~C(비료유형)*C(물주기)',data=df).fit()
anova_table = sm.stats.anova_lm(model)
print(anova_table)
```
